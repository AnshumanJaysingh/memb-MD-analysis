```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import loadtxt
import pandas as pd
import scipy
from scipy import signal
from scipy.stats.stats import pearsonr
import seaborn as sn
```

```python
path="/content/drive/MyDrive/DNA_test/"
datafile=path+"AGD_res.dat"

surfdata=loadtxt(datafile)
np.shape(surfdata)
```

```python
ax = sn.heatmap(surfdata, cmap='icefire')
fig = plt.gcf()
fig.set_size_inches(20,8)
fig.set_dpi(100)
ax.set_xlabel("Time (ns)", fontsize=14)
ax.set_ylabel("Residues in DNA", fontsize=14)
#ax.set_yticks(np.arange(len(res[30:69])))
#ax.set_yticklabels(res[30:69], weight="bold")
ax.set_title("AGD_res system", weight="bold")
```
